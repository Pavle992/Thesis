### Text processing 

```python
contentStrings = comment_db.fetch_all(select='about', from_clause='user_social',data_as_dataframe=True)
# # contentStrings.set_index('id', inplace=True)
contentStrings.head(30)
```

```python
from utils.text_processing_utils.stringParsing import StringTokenizer
import string
from nltk.corpus import stopwords

def getNouns(strText, strTokenizer):
    if strText != None:
        strTokenizer.setString(strText)
#         print(strTokenizer.getNounPhrases())
        return strTokenizer.getNounPhrases()
    else:
        return strText   
    
def removePunctuations(str_param):
    str_param = str_param.lower()
    cleanStr = "".join(x for x in str_param if x not in string.punctuation)
    cleanStr = "".join(x for x in cleanStr if x in string.printable)
    return cleanStr
```

```python
tc = StringTokenizer()
```

```python
contentStrings['novi'] = contentStrings['about'].apply(lambda x: getNouns(removePunctuations(x),tc) if x!=None else [])
```

```python
contentStrings['novi'] = contentStrings['novi'].apply(lambda x: [word for word in x if word not in stopwords.words('english')])
```

```python
# contentStrings.head(30)
```

```python
import bisect
# Function for returning best describer of sentence
def chooseBestToken(listTokens, phrasesDict):   
        # Return token with greatest number of occurences
        # among other words in list in whole dataset
        if len(listTokens) == 0:
            return ""
        if len(listTokens) == 1:
            return listTokens[0]
        
        occurences = []
        for el in listTokens:
            occurences.append(phrasesDict[el])
        
        maks = max(occurences)
        
        for el in listTokens:
            if phrasesDict[el] == maks:
                return el
```

```python
nounPhrases = {}

for x in contentStrings['novi']:
    # Adding elements to list
    for el in x:
        if el not in nounPhrases.keys():
            nounPhrases[el] = 1
        else:
            nounPhrases[el] += 1

print("Finished")
# describe text
contentStrings['descriptor'] = contentStrings['novi'].apply(lambda x: chooseBestToken(x, nounPhrases))
```

```python
def getDictPhrases(targetColumn):
    nounPhrases = {}

    for x in targetColumn:
        # Adding elements to list
        for el in x:
            if el not in nounPhrases.keys():
                nounPhrases[el] = 1
            else:
                nounPhrases[el] += 1
    
    return nounPhrases

def getTopPhrases(dictionary, n=6):
    sortedList = sorted(dictionary.items(), key=lambda x: x[1], reverse=True)[:n]
    return [x[0] for x in sortedList]
```

### Email parsing

```python
def getEmailType(email):
    if email == None:
        return email
    
    i = email.index('@')    
    email_type = email[i+1:]
    i = email_type.index('.')
    
    return email_type[:i]
```

```python
user_social[' emails'] = user_social[' emails'].apply(getEmailType)
```

```python
user_social.head(50)
```
